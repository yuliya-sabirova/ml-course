# №7 Зертханалық жұмыс: Полиномдық регрессия, реттеу және кросс-валидация

**Мақсаты:** Негізгі сызықтық регрессия модельдерінің күрделілігін басқару және артық оқытуға қарсы күрес әдістерін қолдану арқылы оларды жақсартуды үйрену.

**Тапсырма:**
Бұл зертханалық жұмыс №6 Зертханалық жұмыстың тікелей жалғасы болып табылады. Сізге бұрын құрған модельдеріңіздің бірін алып, оның сапасын полиномдық белгілер мен реттеуді қолдану арқылы жақсартуға тырысу керек.

**№6 Зертханалық жұмыста жұмыс істеген нұсқаларыңыздың бірін таңдаңыз (Калифорния, E-commerce немесе Медицина).**

**Қадамдар:**
1.  **Дайындық:** №6 Зертханалық жұмыстан деректер жиынтығын жүктеу және негізгі дайындау ( `X` пен `y` құру, `train_test_split`) үшін кодты көшіріңіз.
2.  **Оңтайлы күрделілікті іздеу:**
    *   Полином дәрежесін 1-ден, мысалы, 5-ке дейін қайталайтын цикл жазыңыз.
    *   Цикл ішінде `StandardScaler`, `PolynomialFeatures` және `LinearRegression`-дан тұратын конвейер (`pipeline`) құрыңыз.
    *   Модельді **оқыту** іріктемесінде оқытыңыз.
    *   RMSE қатесін **оқыту** және **тест** (немесе валидациялық) іріктемелерінде есептеңіз.
    *   Оқыту қисықтарының графигін (RMSE-нің полином дәрежесіне тәуелділігін train және test үшін) құрыңыз. График бойынша оңтайлы полином дәрежесін анықтаңыз.
3.  **Реттеу және гиперпараметрлерді таңдау:**
    *   Алдыңғы қадамнан ең жақсы полином дәрежесін таңдаңыз.
    *   Жаңа конвейер құрыңыз, бірақ енді `LinearRegression` орнына `ElasticNet` қолданыңыз.
    *   `ElasticNet` моделінің `alpha` және `l1_ratio` үшін гиперпараметрлер торын (`param_grid`) құрыңыз.
    *   **Оқыту іріктемесінде** ең жақсы гиперпараметрлер комбинациясын табу үшін `cv=5` (немесе 10) бар `GridSearchCV` қолданыңыз.
4.  **Қорытынды бағалау:**
    *   `GridSearchCV` тапқан ең жақсы модельді алыңыз (`grid_search.best_estimator_`).
    *   Оның соңғы сапасын (RMSE, MAE, R²) **тест іріктемесінде** бағалаңыз.
    *   Алынған метрикаларды №6 Зертханалық жұмыстағы қарапайым сызықтық регрессия метрикаларымен салыстырыңыз. Қорытынды жасаңыз: модельді жақсарту мүмкін болды ма?
