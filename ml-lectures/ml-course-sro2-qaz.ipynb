# "Машиналық оқыту" курсы бойынша жобалық жұмыс
**Тақырыбы: "Машиналық оқыту модельдерін салыстырмалы талдау"**

**Форматы:** 3-4 адамнан тұратын топтардағы білім алушының өздік жұмысы (БӨЖ).

**Максималды балл:** 20.

## 1. Жалпы ақпарат

Сіздер деректерді талдау құралдарын ғана емес, сонымен қатар машиналық оқытудың негізгі алгоритмдерін де меңгердіңіздер. Енді барлық алған білімдеріңізді толыққанды есепті шешу үшін қолданатын уақыт келді.

Сіздер командада ML-зерттеудің толық циклін жүргізулеріңіз керек: деректерді дайындаудан бастап, бірнеше модельді оқытуға, баптауға және салыстыруға дейін. Жұмысыңыздың нәтижесі Jupyter Notebook форматындағы есеп және жобаны ауызша қорғау болады.

**Жобаның мақсаты** — модельдерді жай ғана оқыту емес, оларға салыстырмалы талдау жүргізу, қойылған есепті шешу үшін ең жақсысын таңдау және өз таңдауыңызды сенімді түрде негіздеу.

## 2. Рөлдер және міндеттерді бөлу

Жобамен жұмыс үш кезеңге бөлінеді. Жауапкершіліктің бөлінуіне қарамастан, **әрбір қатысушы бүкіл жобаны толық түсінуге міндетті** және оның кез келген бөлігі бойынша сұрақтарға жауап беруге дайын болуы керек.

#### **1-кезең: Деректерді бірлесіп дайындау (Бүкіл команда)**
*   **Команданың барлық қатысушылары** бірлесіп зерттеушілік деректерді талдауды (EDA) жүргізеді және бос орындарды өңдеу мен белгілерді құру бойынша шешімдер қабылдайды.
*   **Қатысушылардың бірі (командада таңдалады)** өз мойнына **Деректер инженері (Data Engineer)** рөлін алады. Ол кейіннен барлығы қолданатын алдын ала өңдеуге арналған `Pipeline`-ның соңғы, таза кодын жазуға жауапты болады.

---

#### **2-кезең: Жеке модельдеу**
Әрбір қатысушы **өз модельдер жинағын** оқытуға, баптауға (`GridSearchCV`) және бастапқы талдауға жауапкершілік алады.

### **Егер командада 3 адам болса:**

#### **РЕГРЕССИЯ** нұсқалары үшін (1 және 2-нұсқа):

*   **1-қатысушы:**
    1.  `LinearRegression`
    2.  `Ridge` (`alpha` таңдау арқылы)
    3.  `Lasso` (`alpha` таңдау арқылы)

*   **2-қатысушы:**
    1.  `DecisionTreeRegressor` (`max_depth` таңдау арқылы)
    2.  `RandomForestRegressor` (`n_estimators` таңдау арқылы)

*   **3-қатысушы (Топ жетекшісі):**
    1.  `SVR` (Support Vector Regressor, `kernel`, `C`, `gamma` таңдау арқылы)
    2.  `KNeighborsRegressor` (`n_neighbors` таңдау арқылы)

#### **КЛАССИФИКАЦИЯ** нұсқалары үшін (3 және 4-нұсқа):

*   **1-қатысушы:**
    1.  `LogisticRegression`
    2.  `KNeighborsClassifier` (`n_neighbors` таңдау арқылы)

*   **2-қатысушы:**
    1.  `DecisionTreeClassifier` (`max_depth` таңдау арқылы)
    2.  `RandomForestClassifier` (`n_estimators` таңдау арқылы)

*   **3-қатысушы (Топ жетекшісі):**
    1.  `SVC` (Support Vector Classifier, `kernel`, `C`, `gamma` таңдау арқылы)

---
### **Егер командада 4 адам болса:**

#### **РЕГРЕССИЯ** нұсқалары үшін (1 және 2-нұсқа):

*   **1-қатысушы (Data Engineer):** **1-кезеңге**, сондай-ақ `LinearRegression` моделіне жауапты.
*   **2-қатысушы:** `Ridge` және `Lasso`.
*   **3-қатысушы:** `DecisionTreeRegressor` және `KNeighborsRegressor`.
*   **4-қатысушы (Топ жетекшісі):** `RandomForestRegressor` және `SVR`.

#### **КЛАССИФИКАЦИЯ** нұсқалары үшін (3 және 4-нұсқа):

*   **1-қатысушы (Data Engineer):** **1-кезеңге**, сондай-ақ `LogisticRegression` моделіне жауапты.
*   **2-қатысушы:** `KNeighborsClassifier` және `DecisionTreeClassifier`.
*   **3-қатысушы:** `RandomForestClassifier`.
*   **4-қатысушы (Топ жетекшісі):** `SVC`.

#### **3-кезең: Бірлескен талдау және қорытындылар (Бүкіл команда)**
*   **Команданың барлық қатысушылары** бірлесіп алынған метрикаларды талдайды, модельдерді салыстырады және неліктен кейбір алгоритмдердің басқаларына қарағанда жақсырақ жұмыс істегенін талқылайды.
*   **Топ жетекшісі** команданың ортақ пікірін көрсететін қорытынды жазуға жауапты болады.

## 3. Жұмыс жоспары

1.  **Нұсқаны таңдау және деректермен танысу:**
    *   Ұсынылған нұсқалардың бірін таңдаңыз.
    *   **Зерттеушілік деректерді талдау (EDA)** жүргізіңіз. Үлестірімдерді зерттеңіз, ауытқуларды табыңыз, өзара байланыстарды іздеу үшін графиктер құрыңыз. Деректер туралы алдын ала қорытындылар жасаңыз.

2.  **Модельдеу үшін деректерді дайындау:**
    *   Деректерді тазалауды жүргізіңіз (бос орындарды, шығарындыларды өңдеу).
    *   Категориялық белгілерді кодтаңыз (`get_dummies`).
    *   Деректерді оқыту және тест таңдамаларына бөліңіз (`train_test_split`).
    *   Сандық белгілерді масштабтау үшін `Pipeline` дайындаңыз (`StandardScaler`).

3.  **Модельдерді оқыту және баптау:**
    *   Жауапкершілік тізіміңіздегі **әрбір** модель үшін `GridSearchCV` көмегімен гиперпараметрлерді таңдауды жүргізіңіз.
    *   **Маңызды:** Деректердің жылыстауын болдырмау үшін `Pipeline` мен модельді біріктіріңіз.
    *   Әрбір модель үшін табылған ең жақсы параметрлерді шығарыңыз.

4.  **Бағалау және салыстырмалы талдау:**
    *   Барлық модельдердің тест таңдамасындағы болжамдарын жинаңыз.
    *   Сапа метрикаларын есептеп, шығарыңыз. **Регрессия үшін:** R2, MAE, RMSE. **Классификация үшін:** `classification_report`.
    *   Барлық оқытылған модельдердің негізгі метрикаларын көрнекі түрде салыстыратын **жинақ кестесін немесе графигін** жасаңыз.
    *   **Ең жақсы модельді таңдап** және жазбаша **өз таңдауыңызды негіздеңіз**: сіздің ойыңызша, деректер құрылымын және алынған метрикаларды ескере отырып, неліктен дәл осы алгоритм есепті ең жақсы орындады?

5.  **Ең жақсы модельді терең талдау:**
    *   Ең жақсы моделіңізді егжей-тегжейлі талдаңыз: коэффициенттерге (сызықтық модельдер үшін) немесе белгілердің маңыздылығына (`feature_importances_` ормандар үшін) қараңыз. Нысаналы айнымалыға қандай факторлар ең қатты әсер ететіні туралы қорытынды жасаңыз.

*   **Нысаналы айнымалы:** `quality` (3-тен 8-ге дейінгі баға).
*   **Есеп:** Қызыл шараптың физика-химиялық көрсеткіштері бойынша оның сапа бағасын болжау. Бұл **көп класты классификация** есебі. Нақты жобаларда есепті кейде бинарлық классификацияға («сапалы / сапасыз») келтіреді, бірақ осы тапсырма аясында көп класты классификацияны іске асырамыз.
*   **Маңызды:** Бұл деректер жинағындағы класстар **теңгерімсіз**. Модельдерді бағалау кезінде `accuracy`-ге емес, `f1-score (weighted)` немесе `f1-score (macro)`-ға назар аударыңыз.

## 5. Есептің құрылымы (Jupyter Notebook)

Есеп Markdown-ұяшықтары арқылы құрылымдалған бірыңғай `.ipynb` файлы болуы керек.

1.  **Титул парағы:**
    *   Жобаның атауы, нұсқа нөмірі.

4.  **Деректерді дайындау:**
    *   Деректерді тазалау, кодтау және бөлу бойынша әрекеттеріңізді түсіндіретін код және түсініктемелер.

    *   Әрбір модель үшін өз бөлімшесі болуы керек.

6.  **Салыстырмалы талдау және ең жақсы модельді таңдау:**
    *   Барлық модельдердің метрикалары бар жинақ кестесі/графигі.
    *   Ең жақсы модельді таңдауды мәтіндік негіздеу.
7.  **Ең жақсы модельді талдау:**
    *   Белгілердің маңыздылығын немесе коэффициенттерді визуализациялау және интерпретациялау.
8.  **Жоба бойынша жалпы қорытындылар:**
    *   Жұмыс барысында алынған қысқаша түйіндеме (3-5 негізгі инсайт). Қандай факторлар ең маңызды болды? Қай модель жеңді және неліктен?
9.  **Қатысушылардың үлесі:**
