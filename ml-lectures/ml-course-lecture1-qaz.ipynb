# №1 Дәріс: Машиналық оқытуға кіріспе

**Курс:** Машиналық оқыту

---

### Дәрістің мақсаттары

Пәннің біртұтас көрінісін қалыптастыру және үш негізгі сұраққа жауап беру:

1. **Не** интуитивті, тарихи және математикалық тұрғыдан машиналық оқыту дегеніміз не?
2. **Неліктен** ол қажет және қандай нақты мәселелерді шешеді?
3. **Қалай** машиналық оқыту саласындағы типтік жоба құрылады және біздің курсымыздың жол картасы қандай?

Бұл дәріс — Data Science әлемінде бағдар табуға көмектесетін және алдағы апталарда әртүрлі құралдарды не үшін зерттейтінімізді түсінуге жәрдемдесетін "жергілікті карта".

## 1. Машиналық оқыту (Machine Learning) дегеніміз не?

Күн сайын сіз негізінде машиналық оқыту жатқан ондаған жүйелермен бетпе-бет келесіз. Netflix сізге фильм ұсынады, пошта клиенті хатты спамға жібереді, телефоныңыз сізді бетіңізден таниды, ал навигатор оңтайлы маршрутты құрады.

Классикалық бағдарламалауда біз компьютердің мәселені шешуі үшін нақты ережелер жиынтығын (алгоритм) жазамыз. Бірақ ережелер тым көп болса немесе біз оларды өзіміз тұжырымдай алмасақ не істейміз? Мысалы, фотосуреттегі мысықты иттен қалай ажыратуға болатын ережелерді қалай жазуға болады?

> **Машиналық оқытудың (ML) негізгі идеясы қарапайым:** бағдарламашы миллиондаған қатаң `if-then-else` ережелерін қолмен жазудың орнына, біз компьютерге мыңдаған мысалдарды көрсетеміз, және ол **деректердегі жасырын заңдылықтарды өзі табады**.

### Қысқаша тарихы және анықтамалар

ML бумы соңғы онжылдықта болғанымен, идеяның өзі жаңа емес.

* **1959 жыл, Артур Сэмюэл**, жасанды интеллект саласының пионері, дойбы ойнайтын бағдарламамен жұмыс істеп, классикалық анықтама берді:
 > *"Машиналық оқыту — компьютерлерге нақты бағдарламалаусыз оқу қабілетін беретін зерттеу саласы".*
 Оның бағдарламасы өзімен мыңдаған партия ойнап, кез келген адамнан мықтырақ болды.

* **1997 жыл, Том Митчелл**, формалды, заманауи анықтама берді:
 > *"Егер компьютерлік бағдарламаның белгілі бір **T тапсырмасын** шешудегі **P өнімділігі** **E тәжірибесіне** сүйеніп өлшенгенде, E тәжірибесін жинақтаған сайын жақсарса, онда ол E тәжірибесінен үйренеді дейді".*
 * **Тапсырма (T):** Дойбы ойнау.
 * **Тәжірибе (E):** Бағдарлама ойнаған мыңдаған партиялар.
 * **Өнімділік (P):** Жеңген партиялардың пайызы.

## 2. ML жобасының өмірлік циклі

Алгоритмдерге тереңдемес бұрын, ML жобасының қандай кезеңдерден тұратынын түсіну маңызды. Барлығы нақты әлемнен басталып, қолданбалы шешіммен аяқталады.

```mermaid
graph TD;
 A[Нақты әлем] --> B(Шикі деректерді жинау);
 B --> C(Өңдеу және сақтау);
 subgraph Data Engineering
 B; C;
 end
 C --> D(Тазалау және құрылымдау);
 D --> E(Деректерді барлау талдауы, EDA);
 subgraph Data Analysis
 D; E;
 end
 E --> F(ML модельдерін құру);
 subgraph Data Science / ML Engineering
 F
 end
 F --> G[Деректерге негізделген өнім];
 F --> H[Талдамалық есептер мен дашбордтар];
 G --> A;
 H --> A;
```

1. **Data Engineering:** Модельдеріміз үшін "жанармай" жинау және дайындау кезеңі. Бұл дерекқорлармен, сенсорлармен, логтармен жұмыс.
2. **Data Analysis:** Талдаушылар мен деректер ғалымдары деректерді түсінуге тырысатын кезең: қоқыстан тазалайды, аномалияларды іздейді, алғашқы гипотезаларды табу үшін визуализациялар жасайды.
3. **Machine Learning:** Дайындалған деректер негізінде болжамдық модель құрылады және оқытылады.
4. **Енгізу (Внедрение):** Модель өнімге (мысалы, мобильді қосымшаға) енгізіледі немесе оның нәтижелері бизнес-шешімдер қабылдау үшін (мысалы, есептер арқылы) қолданылады.

Біздің курс **Data Analysis** және **Machine Learning** кезеңдеріне бағытталған.

## 3. Машиналық оқыту тілі: Негізгі ұғымдар

Енді интуицияны қатаң тілге аударайық. Алдымен терминдер деңгейінде, содан кейін — математика деңгейінде. Қарапайым мысал алайық: біз пәтер бағасын болжауды үйренгіміз келеді.

| Термин | Сипаттамасы | Мысал (пәтер бағасын болжау) |
|:---|:---|:---|
| **Нысан (Объект, `x`)** | Біз зерттеп, болжам жасайтын нәрсе. | Нақты пәтер |
| **Белгілер (Признаки, `features`)**| Вектор түрінде ұсынылған нысанның өлшенетін сипаттамалары. | `(Аудан, Бөлмелер саны, Қабат, Метроға дейінгі қашықтық)` |
| **Жауап (Ответ, `y`)** | Біз болжағымыз келетін мақсатты айнымалы. | Пәтер бағасы |
| **Оқыту үлгісі (Обучающая выборка)**| Көптеген нысандардың белгілері мен дұрыс жауаптары белгілі деректер жиынтығы `{x_i, y_i}`. | 1000 сатылған пәтердің сипаттамалары мен бағалары бар кесте |

### Машиналық оқыту есебінің оптимизациялық қойылымы

Енді оқыту есебін математикалық тұрғыдан формализациялайық. Бұл слайд біз зерттейтін барлық алгоритмдердің негізі болып табылады.

**Берілгені:**
* $X$ — нысандар кеңістігі (белгілермен сипатталады).
* $Y$ — рұқсат етілген жауаптар жиыны (бағалар, класстар, т.б.).
* Белгісіз мақсатты тәуелділік $y(x): X \to Y$ бар. Біз оның түрін білмейміз, бірақ бізде оның нақты мысалдардағы көріністері бар.
* Оқыту үлгісі (training sample) $\{x_1, \dots, x_\ell\} \subset X$ — дұрыс жауаптары белгілі нысандардың шектеулі саны.
* Біз тәуелділіктің түрі туралы гипотеза ұсынамыз, **параметрлік модельді** $a(x, w)$ таңдаймыз, мұндағы $w$ — бапталатын параметрлер (салмақтар) векторы.

**Табу керек:**
* Біздің $a(x, w)$ моделіміз шынайы $y(x)$ тәуелділігіне мүмкіндігінше жақын болатындай $w$ параметрлер векторын табу.

**Критерий (Қалай табуға болады?):**
* Біз бір нысандағы $x$ алгоритміміздің қаншалықты қателескенін көрсететін **шығын функциясын (loss function)** $L(w, x)$ енгіземіз. $a(x, w)$ болжамының дұрыс жауап $y(x)$-тен ауытқуы неғұрлым көп болса, $L$ мәні соғұрлым үлкен болады.
* Біз **эмпирикалық тәуекелді (empirical risk)** — оқыту үлгісінің барлық нысандарындағы орташа қатені — минимизациялаймыз:
$$ \sum_{i=1}^{\ell} L(w, x_i) \to \min_{w} $$

> **Қарапайым тілмен айтқанда:** Модельді оқыту — бұл оптимизация есебі. Біз оқыту деректеріндегі қателіктер үшін жалпы айыппұл минималды болатындай `w` параметрлерін іздейміз.

### ML-дегі есептердің түрлері

$Y$ жауаптар жиынының түріне байланысты есептер бірнеше негізгі түрге бөлінеді:

#### Мұғаліммен оқыту (Supervised Learning)
Бізде "мұғалім" бар — белгілі дұрыс жауаптары бар оқыту үлгісі.

* **Классификация есебі (Задача Классификации):** $y$ жауаптары — шектеулі жиынтықтан алынған белгілер (класстар). Мысалы, $Y = \{\text{иә}, \text{жоқ}\}$ немесе $Y = \{\text{мысық}, \text{ит}, \text{құс}\}$.
* **Регрессия есебі (Задача Регрессии):** $y$ жауабы — нақты сан. Мысалы, $Y = \mathbb{R}$.

#### Мұғалімсіз оқыту (Unsupervised Learning)
"Дұрыс жауаптар" жоқ, тек $x_i$ нысандары бар.

* **Кластерлеу есебі (Задача Кластеризации):** Ұқсас нысандарды топтарға біріктіру керек, бұл топтардың не екенін алдын ала білмейміз.

Басқа да күрделі түрлері бар, мысалы, нығайтумен оқыту (Reinforcement Learning), бірақ біздің курс аясында осы үшеуіне назар аударамыз.

### Қолданбалы есептердің мысалдары

1. **Классификация есебі: Кредиттік скоринг**
 * **Нысан:** Несие алуға өтініш берген қарыз алушы.
 * **Белгілер:** Жасы, табысы, жұмыс өтілі, білімі, басқа несиелердің болуы.
 * **Жауап (класс):** Несиені қайтарады ("жақсы" клиент) немесе қайтармайды ("жаман" клиент).
 * **Мақсат:** Банктің тәуекелдерін азайта отырып, өтінімдерді автоматты түрде мақұлдайтын немесе қабылдамайтын модель құру.

2. **Регрессия есебі: Тұтынушылық сұранысты болжау**
 * **Нысан:** Белгілі бір дүкендегі нақты тауар.
 * **Белгілер:** Апта күні, тауар бағасы, жарнамалық акциялардың болуы, маусым, ауа-райы.
 * **Жауап (сан):** Келесі күні сатылған тауар бірліктерінің саны.
 * **Мақсат:** Сөрелерде тауар тапшылығы немесе артық қалдық болмас үшін сатып алулар мен логистиканы оңтайландыру.

3. **Кластерлеу есебі: Клиенттерді сегменттеу**
 * **Нысан:** Телекоммуникациялық компанияның клиенті.
 * **Белгілер:** Қаншалықты жиі қоңырау шалады, интернетке қанша жұмсайды, қандай қосымша қызметтерді пайдаланады, қанша уақыттан бері клиент болып табылады.
 * **Жауап:** Белгісіз.
 * **Мақсат:** Әр топқа жеке тарифтер мен акциялар ұсыну үшін барлық клиенттерді ұқсас мінез-құлықтары бойынша топтарға (сегменттерге) бөлу.

## 4. Басты мәселе: Артық оқыту (Overfitting)

> **Аналогия:** Емтиханға дайындалып жатқан студентті елестетіңіз. Ол есептерді шешудің жалпы принциптерін түсіне алады немесе дайындық үшін берілген 50 билеттің жауаптарын жаттап ала алады. Осы 50 билет бойынша ол тамаша нәтиже көрсетеді. Бірақ емтиханда оған шарты сәл өзгертілген 51-билет берілсе не болады? Ол құлап қалады.

Машиналық оқыту модельдерімен де солай болады. Тым күрделі модель **жалпы заңдылықтарды** іздеудің орнына, оқыту деректерін барлық кездейсоқ шуларымен қоса "жаттап алады". Мұндай модель оқыту үлгісінде өте жақсы жұмыс істейді, бірақ жаңа деректерде мүлдем пайдасыз болады.

### Артық оқытумен қалай күресуге болады?

Шешімі қарапайым: біз ешқашан модельдің сапасын оны оқытқан деректерде тексермейміз.

**Стандартты тәсіл:** Біздің барлық бастапқы деректер базамыз екі (немесе үш) бөлікке бөлінеді:

1. **Оқыту үлгісі (Train set):** Модельдің $w$ параметрлерін баптау үшін қолданылады (яғни, эмпирикалық тәуекелді минимизациялау үшін).
2. **Тест үлгісі (Test set):** Модель сапасын соңғы, әділ бағалау үшін қолданылады. Бұл деректерді модель бірінші рет "көреді".

> **Негізгі тезис:** Модельдің шынайы сапасы оның **жалпылау қабілетімен**, яғни жаңа, бұрын көрмеген деректерде жақсы жұмыс істеуімен анықталады.

## 5. Біздің іс-қимыл жоспарымыз: Курстың құрылымы

Енді бізде жалпы көрініс бар болғандықтан, курсымыздың қалай құрылатынын қарастырайық. Ол екі үлкен логикалық блоктан тұрады.

### 1-модуль: Құрал-саймандар (Алғашқы 5 тақырып)

ML-модельдерін құру үшін бізге қуатты және ыңғайлы "жұмыс үстелі" қажет. Біз үшін мұндай үстел **Python** тілі және оның деректерді талдауға арналған негізгі кітапханалары болады. Алғашқы апталарды мыналарды меңгеруге арнаймыз:

* **Python & Jupyter:** Синтаксис негіздері және интерактивті орта.
* **NumPy:** Жылдам математикалық (векторлық) есептеулер үшін.
* **Pandas:** Кестелік деректермен (біздің "оқыту үлгілеріміз") жұмыс істеу үшін.
* **Matplotlib / Seaborn:** Деректерді "көру" және барлау талдауы үшін әдемі графиктер салу үшін.

Бұл одан әрі қарай жылжу мүмкін емес қажетті іргетас.

### 2 және 3-модульдер: Машиналық оқыту алгоритмдері мен әдістері

Құрал-саймандарды меңгерген соң, біз модельдердің өздерін қарапайымнан күрделіге қарай зерттеуге көшеміз:

* **Сызықтық модельдер** (Сызықтық және Логистикалық регрессия)
* **Сапа метрикалары және модельдерді бағалау әдістері** (Кросс-валидация)
* **Деректерді дайындау** (Белгілер инженериясы)
* **Метрикалық әдістер** (k-Ең жақын көршілер)
* **Жетілдірілген әдістер:** SVM, Шешімдер ағашы, Ансамбльдер (Random Forest)
* **Мұғалімсіз оқыту** (K-Means кластерлеуі)


### Түйіндеме және келесі қадамдар

* **Бүгін не білдік:**
 * Машиналық оқыту — бұл болашақты болжауға арналған модельдер құру үшін деректердегі заңдылықтарды іздеу.
 * Оқыту — бұл жалпы қатені (эмпирикалық тәуекелді) минимизациялайтын оптимизация есебі.
 * Негізгі есеп түрлері: классификация, регрессия және кластерлеу.
 * ML-дегі басты мәселе — артық оқыту, онымен күресу үшін деректерді оқыту және тест үлгілеріне бөледі.

* **Әрі қарай не болады?**
 * **Бүгінгі семинар мен зертханалық жұмыста** біз жұмыс ортасын (Jupyter Notebook) баптап, негізгі жұмыс құралымыз — Python негіздеріне кірісеміз.
