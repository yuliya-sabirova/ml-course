# "Машиналық оқытуға кіріспе" курсы бойынша жобалық жұмыс
**Тақырыбы: "Деректерді барлау талдауы (EDA)"**

**Форматы:** 3 адамнан тұратын топтардағы білім алушының өздік жұмысы (БӨЖ).

**Ең жоғары балл:** 20.

## 1. Жалпы ақпарат

Соңғы 5 аптада сіз Python-да деректермен жұмыс істеуге арналған негізгі құралдарды меңгердіңіз: NumPy, Pandas, Matplotlib және Seaborn. Енді осы білімді іс жүзінде қолданатын уақыт келді.

Сіз үш адамнан тұратын командада ұсынылған деректер жиынтықтарының біріне деректерді барлау талдауының (EDA) толық циклін жүргізуіңіз керек. Жұмысыңыздың нәтижесі Jupyter Notebook форматындағы есеп және жобаны ауызша қорғау болады.

**Жобаның мақсаты** — мүмкіндігінше көп графиктер салу ғана емес, деректерге дұрыс сұрақтар қою, визуализация көмегімен оларға жауаптар табу және ұсынылған зерттеу жоспарына сәйкес мағыналы қорытындылар жасау.

## 2. Жобаның мақсаттары мен міндеттері

*   **Басты мақсат:** Деректердегі негізгі заңдылықтарды, өзара байланыстарды және ауытқуларды анықтау үшін деректерге кешенді барлау талдауын жүргізу.
*   **Міндеттер:**
    1.  Талдау үшін деректерді жүктеу, тазалау және дайындау.
    2.  Өз нұсқаңыз үшін **"Міндетті талдау бағдарламасын"** орындау.
    3.  Атқарылған жұмыс негізінде қорытынды тұжырымдар жасау.
    4.  Түсінікті есеп (Jupyter Notebook) және қорғау үшін 7-10 минуттық презентация дайындау.

## 3. Командадағы рөлдер

Тиімді жұмыс істеу және әділ жеке бағалау үшін әр команда келесі рөлдерді бөлуі керек. Мамандануға қарамастан, **әрбір қатысушы бүкіл жобаны толық түсінуге** және оның кез келген бөлігі бойынша сұрақтарға жауап беруге дайын болуға міндетті.

### 1. Деректер инженері (Data Engineer)
*   **Жауапкершілігі:** Деректерді талдауға жүктеу, тазалау және дайындау.
*   **Міндеттері:**
    *   Деректерді жүктеуге арналған код жазу.
    *   Бастапқы талдау жүргізу: `.info()`, `.describe()`, `.isNone().sum()`.
    *   Жетіспейтін мәндер мен дубликаттарды өңдеу стратегиясын әзірлеу және қолдану.
    *   Деректер түрлерін тексеру және түзету.
    *   Талдау үшін пайдалы болуы мүмкін 1-2 жаңа белгі (feature engineering) жасау (мысалы, күннен жылды алу, сирек санаттарды топтастыру).
*   **Есепте:** "Деректерді жүктеу және алдын ала өңдеу" бөліміне жауап береді.

### 2. Деректер талдаушысы (Data Analyst)
*   **Жауапкершілігі:** Деректердің негізгі сипаттамаларын және тікелей өзара байланыстарын зерттеу.
*   **Міндеттері:**
    *   **Бір өлшемді талдау:** Негізгі сандық белгілердің таралуын (`histplot`, `boxplot`) және санаттық белгілердің жиілігін (`countplot`) зерттеу.
    *   **Екі өлшемді талдау:** Белгілер жұптары арасындағы тікелей байланыстарды табу және визуализациялау: "сан-сан" (`scatterplot`), "санат-сан" (`boxplot`, `barplot`).
    *   Әрбір салынған график бойынша қорытынды жасау.
*   **Есепте:** "Бір өлшемді талдау" және "Екі өлшемді талдау" бөлімдеріне жауап береді.

### 3. Жетекші талдаушы / Топ жетекшісі (Lead Analyst / Team Lead)
*   **Жауапкершілігі:** Өзара байланыстарды терең талдау, барлық нәтижелерді синтездеу және қорытынды көріністі қалыптастыру.
*   **Міндеттері:**
    *   **Корреляциялық талдау:** Өзара байланыстардың жалпы көрінісін бағалау үшін барлық сандық белгілерге арналған корреляциялардың жылу картасын (`heatmap`) құру. **Сізге `sns.heatmap()` синтаксисін өз бетіңізше зерттеу қажет болады**.
    *   **Көп өлшемді талдау:** Талдаудың алдыңғы кезеңдерінде пайда болған гипотезаларды тексеру үшін 3 немесе одан да көп айнымалыны пайдалана отырып, күрделі графиктерді (мысалы, `hue` параметрі бар `scatterplot` немесе `boxplot`) құру.
    *   **Синтез және қорытындылар:** Бүкіл команданың тапқандарын біртұтас баяндауға біріктіру. Зерттеудің басты сұрақтарына жауап беретін соңғы қорытындыны жазу.
    *   **Үйлестіру:** Есептің барлық бөліктерінің бір-бірімен логикалық байланысты екеніне көз жеткізу.
*   **Есепте:** "Көп өлшемді талдау" және "Жалпы қорытындылар" бөлімдеріне жауап береді.

## 4. Тапсырма нұсқалары және міндетті талдау бағдарламасы

Әр команда келесі нұсқалардың бірін алады. Сіздің есебіңізде нұсқаңызға арналған талдау бағдарламасының **барлық** тармақтары бойынша жауаптар мен визуализациялар болуы керек.

### **1-нұсқа: Жылжымайтын мүлік нарығын талдау**
*   **Деректер жиынтығы:** [Kaggle: House Prices](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data) (`train.csv` файлын пайдалану)
*   **Мақсатты айнымалы:** `SalePrice`

**Міндетті талдау бағдарламасы:**
1.  **Нарықтың жалпы жағдайы қандай?** Үй бағасының таралуын (`SalePrice`) талдаңыз. Оны сипаттаңыз: ол симметриялы ма? Медианалық және орташа бағалар қандай? Айқын ауытқулар бар ма?
2.  **Не маңызды: көлемі ме, әлде сапасы ма?** Тұрғын ауданының (`GrLivArea`) және жалпы сапасының (`OverallQual`) сату бағасына әсерін салыстырыңыз. Осы екі фактордың қайсысы, сіздің ойыңызша, жоғары бағаның күштірек предикторы болып табылады? Өз тұжырымдарыңызды графиктермен растаңыз.
3.  **Қайда салу керек: аудандарды талдау.** Ең қымбат 3 және ең арзан 3 ауданды (`Neighborhood`) анықтаңыз. Осы аудандардағы үйлер орташа сапасы (`OverallQual`) және салынған жылы (`YearBuilt`) бойынша ерекшелене ме, соны талдаңыз.
4.  **Жаңа ма, әлде ескі ме?** Салынған жылы (`YearBuilt`) мен қайта жөндеу жылының (`YearRemodAdd`) құнға қалай әсер ететінін зерттеңіз. Ескі, бірақ жөнделген үйлердің бағасында жай ескі үйлермен салыстырғанда айтарлықтай айырмашылық бар ма?
5.  **Қосымша ыңғайлылықтардың рөлі.** Гараж көлемінің (`GarageCars`) және бассейннің болуы/сапасының (`PoolQC`) бағаға әсерін бағалаңыз. Бұл ыңғайлылықтар шешуші факторлар болып табыла ма?
6.  **Сәулеттік ерекшеліктер.** Үйлердің қандай стильдері (`HouseStyle`) ең көп таралған? Әртүрлі стильдердің медианалық бағаларын салыстырыңыз. Басқаларына қарағанда тұрақты түрде қымбат сатылатын стиль бар ма?
7.  **Жасырын байланыстарды іздеу.** Сіз ең маңызды деп санайтын 10 сандық белгіге (соның ішінде `SalePrice`) арналған корреляциялардың жылу картасын құрыңыз. 2-3 ең күшті және күтпеген корреляцияны тауып, сипаттаңыз.
8.  **Қорытынды портрет.** Эймс қаласында максималды инвестициялық тартымдылығы бар үй сатып алғысы келетін адамға арналған 3-4 негізгі ұсыныс жасаңыз. Ол бірінші кезекте қандай сипаттамаларға назар аударуы керек?

### **2-нұсқа: Телеком-компания клиенттерінің кетуін талдау**
*   **Деректер жиынтығы:** [Kaggle: Telco Customer Churn](https://www.kaggle.com/blastchar/telco-customer-churn)
*   **Мақсатты айнымалы:** `Churn`

**Міндетті талдау бағдарламасы:**
1.  **Мәселенің ауқымы қандай?** `Churn` айнымалысындағы кластардың балансын талдаңыз. Кетудің нақты пайызын есептеп, таңдаманың теңгерімсіз екендігі туралы қорытынды жасаңыз.
2.  **"Кіріптарлық" келісімшарттар.** Әртүрлі келісімшарт түрлері (`Contract`) үшін кету деңгейін салыстырыңыз. Клиенттер әрбір келісімшарт түрімен орташа есеппен қанша уақыт (`tenure`) болады? Компания үшін ең сенімді келісімшарт қайсысы?
3.  **Баға мәселесі.** Ай сайынғы төлемдер (`MonthlyCharges`) мен кету арасындағы байланысты зерттеңіз. Клиенттер жиі кете бастайтын бағада "ауыру шегі" бар ма? Талдау үшін кем дегенде екі түрлі график түрін пайдаланыңыз.
4.  **Демография және адалдық.** Демографиялық факторлардың (`gender`, `SeniorCitizen`, `Partner`, `Dependents`) кетуге әсер ететінін талдаңыз. Ең адал демографиялық топтың портретін жасаңыз.
5.  **Интернет — екі ұшы бар таяқ па?** Интернет-қызмет түрі (`InternetService`) тұрғысынан кетуді зерттеңіз. Белгілі бір қосымша қызметтердің (`OnlineSecurity`, `TechSupport`) болуы кетудің төмендеуімен байланысты ма?
6.  **Ыңғайлылық па, әлде әдет пе?** Төлем әдісі (`PaymentMethod`) адалдыққа әсер ете ме? Кететін клиенттер таңдайтын төлем әдісі бар ма?
7.  **Өзара байланыстарды іздеу.** Сандық белгілер үшін корреляциялардың жылу картасын құрыңыз. Клиенттің "өмір сүру ұзақтығы", оның ай сайынғы және жалпы төлемдері арасында күшті байланыс бар ма?
8.  **"Кету алдындағы клиент" профилі.** Келесі айда кету ықтималдығы жоғары клиенттің 3-4 негізгі белгісін сипаттаңыз. Сіздің қорытындыларыңыз негізінде компания қандай алдын алу шараларын қабылдай алар еді?

### **3-нұсқа: Бейнеойындарды зерттеу**
*   **Деректер жиынтығы:** [Kaggle: Video Game Sales](https://www.kaggle.com/gregorut/videogamesales)
*   **Мақсатты айнымалы:** `Global_Sales`

**Міндетті талдау бағдарламасы:**
1.  **Нарықты кім басқарады?** Жалпы әлемдік сатылымдар (`Global_Sales`) бойынша топ-5 платформаны (`Platform`), жанрды (`Genre`) және баспагерді (`Publisher`) анықтаңыз.
2.  **Индустрияның эволюциясы.** Жылдар бойынша (`Year`) сатылымдардың жалпы көлемі мен шығарылатын ойындар санының динамикасын көрсететін график құрыңыз. "Алтын дәуірді", құлдырау кезеңін және нарықтың қазіргі жағдайын бөліп көрсетіп, сипаттаңыз.
3.  **Платформалардың өмірлік циклі.** 3-4 танымал платформаны таңдаңыз (мысалы, PS2, X360, Wii). Олардың әрқайсысы үшін жылдар бойынша сатылым динамикасының графигін құрыңыз. Олардың өмірлік циклдерін салыстырыңыз: олар қаншалықты тез танымал болды және қаншалықты тез сөнді?
4.  **Талғамдағы айырмашылық: Батыс пен Шығыс.** Солтүстік Америкадағы (`NA_Sales`) және Жапониядағы (`JP_Sales`) ең көп сатылатын топ-5 жанрды салыстырыңыз. Осы екі аймақтың ойын талғамдарындағы негізгі айырмашылықтарды сипаттаңыз.
5.  **Блокбастер формуласы.** Әлемдегі ең көп сатылатын топ-100 ойынды талдаңыз. Бұл тізімде қандай жанрлар мен платформалар жиі кездеседі? Хит жасаудың ортақ "формуласын" анықтауға бола ма?
6.  **Баспагерлердің мамандануы.** 2-3 ірі баспагерді таңдаңыз (мысалы, Nintendo, Electronic Arts, Activision). Олардың портфолиосын талдаңыз: олар қандай жанрларға маманданған және оларда қаншалықты табысты?
7.  **Нарықтардың корреляциясы.** Солтүстік Америка, Еуропа және Жапония нарықтары бір-бірімен қаншалықты тығыз байланысты? `NA_Sales`, `EU_Sales`, `JP_Sales` үшін корреляциялардың жылу картасын құрыңыз.
8.  **Инвесторға ұсыныстар.** Қазір сіздің деректер жиынтығыңыздағы соңғы жыл деп елестетіңіз. Трендтерге сүйене отырып, инвесторға кеңес беріңіз: ол қандай жанрдағы, қандай платформаға арналған және қай аймаққа бағытталған ойынға инвестиция салуы керек? Жауабыңызды негіздеңіз.

### **4-нұсқа: Жүрек-қан тамырлары ауруларын талдау**
*   **Деректер жиынтығы:** [https://www.kaggle.com/datasets/johnsmith88/heart-disease-dataset)
*   **Мақсатты айнымалы:** `target` (1 — ауру бар, 0 — жоқ)

**Міндетті талдау бағдарламасы:**
1.  **Жалпы көріністі бағалау.** Таңдамадағы сау және науқас пациенттердің (`target`) арақатынасын талдаңыз. Пациенттердің жасы (`age`) және жынысы (`sex`) бойынша таралуын зерттеңіз.
2.  **Жас – қауіп факторы.** Ауру қаупі жасқа қаншалықты қатты байланысты? Әртүрлі жас топтарындағы (мысалы, 30-40, 40-50, 50-60, 60+ жас) науқас пациенттердің үлесін көрсететін график құрыңыз.
3.  **"Үнсіз қанішерлер": қан қысымы және холестерин.** Сау және науқас пациенттердегі артериялық қысымның (`trestbps`) және холестерин деңгейінің (`chol`) таралуын салыстырыңыз. Бұл көрсеткіштер диагностика үшін қаншалықты ақпаратты?
4.  **Симптомдар және диагноз.** Кеудедегі ауырсынудың (`cp`) қандай түрі және ЭКГ (`restecg`) нәтижелері аурудың болуымен тығыз байланысты? Тұжырымдарыңызды визуализациялармен растаңыз.
5.  **Жүктемеге реакция.** Сау және науқас пациенттердегі максималды жүрек соғу жиілігін (`thalach`) және ST сегментінің депрессиясын (`oldpeak`) салыстырыңыз. Ағзаның физикалық жүктемеге реакциясы негізгі көрсеткіш болып табыла ма?
6.  **Гендерлік ерекшеліктер.** Ерлер мен әйелдер арасында негізгі көрсеткіштердің (холестерин, қан қысымы, макс. пульс) орташа мәндерінде айырмашылық бар-жоғын талдаңыз. Жыныс осы көрсеткіштердің аурудың болуымен байланысына әсер ете ме?
7.  **Жасырын өзара байланыстар.** Барлық сандық белгілер үшін корреляциялардың жылу картасын құрыңыз. 2-3 ең күшті немесе ең қызықты корреляцияны тауып, түсіндіріңіз.
8.  **Жоғары қауіп тобындағы пациенттің профилі.** Тапқандарыңызды қорытындылаңыз. Пациентте жүрек ауруының болуын ең жоғары ықтималдылықпен көрсететін 3-4 негізгі сипаттаманы (медициналық көрсеткіштер мен симптомдар) сипаттаңыз.

## 5. Есепке қойылатын талаптар (Jupyter Notebook)

Есеп Markdown ұяшықтарының көмегімен құрылымдалған бірыңғай `.ipynb` файлы болуы керек.
1.  **Титул парағы:** Жобаның атауы, нұсқа нөмірі, рөлдері көрсетілген барлық қатысушылардың аты-жөні.
2.  **Кіріспе:** Деректер жиынтығының және жұмыс мақсатының қысқаша сипаттамасы.
3.  **Деректерді жүктеу және алдын ала өңдеу:** Сіздің әрекеттеріңізді түсіндіретін код және түсініктемелер.
4.  **Деректерді барлау талдауы (EDA):** Бұл бөлім сіздің нұсқаңызға арналған "Міндетті талдау бағдарламасы" тармақтарына сәйкес құрылымдалуы керек. Әрбір тармақта код ұяшықтары және сіздің қорытындыларыңыз бар ұяшықтар болуы керек.
5.  **Жалпы қорытындылар:** Талдау барысында алынған қысқаша түйіндеме (3-5 негізгі инсайт).
6.  **Қатысушылардың үлесі:** Әр қатысушының жобаға қосқан үлесінің қысқаша сипаттамасы.
